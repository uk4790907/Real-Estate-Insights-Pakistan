```python
# location VS price
top_location = Real_Estate_df.groupby('location')['price'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(14,7))
plt.bar(top_location.index, top_location.values)
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.title('Top 10 Locations by Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
Real_Estate_df = Real_Estate_df.dropna(subset=['location', 'price', 'area_combined'])
Real_Estate_df['price_per_marla'] = Real_Estate_df['price'] / Real_Estate_df['area_combined']
```

```python
avg_price_location = Real_Estate_df.groupby('location')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(14,7))
plt.bar(avg_price_location.index, avg_price_location.values)
plt.xticks(rotation=45)
plt.xlabel("Location")
plt.ylabel("Average Price")
plt.title("Top 10 Locations by Average Price")
plt.show()
```

```python
price_per_area = Real_Estate_df.groupby('location')['price_per_marla'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(14,7))
plt.bar(price_per_area.index, price_per_area.values)
plt.xticks(rotation=45)
plt.xlabel("Location")
plt.ylabel("Average Price per Marla")
plt.title("Top 10 Locations by Price per Marla")
plt.grid(True)
plt.show()
```

```python
demand_price = Real_Estate_df.groupby('location').agg(
    total_listings=('location', 'count'),
    avg_price=('price', 'mean')
).reset_index()

# High demand + low price
best_value = demand_price.sort_values(
    by=['total_listings', 'avg_price'],
    ascending=[False, True]
).head(10)

plt.figure(figsize=(14,7))
plt.bar(best_value['location'], best_value['total_listings'])
plt.xticks(rotation=45)
plt.xlabel("Location")
plt.ylabel("Total Listings (Demand)")
plt.title("High Demand & Lower Price Areas")
plt.grid(True)
plt.show()
```

```python
growth = Real_Estate_df.groupby('location').agg(
    avg_price_per_marla=('price_per_marla', 'mean'),
    listings=('location', 'count')
).reset_index()

growth_potential = growth.sort_values(
    by=['avg_price_per_marla', 'listings'],
    ascending=[True, False]
).head(10)

plt.figure(figsize=(14,7))
plt.bar(growth_potential['location'], growth_potential['listings'])
plt.xticks(rotation=45)
plt.xlabel("Location")
plt.ylabel("Listings")
plt.title("Areas with Better Growth Potential")
plt.grid(True)
plt.show()

```

```python
cheap_large = Real_Estate_df.groupby('location').agg(
    avg_price=('price', 'mean'),
    avg_area=('area_combined', 'mean')
).reset_index()

cheap_large = cheap_large.sort_values(
    by=['avg_price', 'avg_area'],
    ascending=[True, False]
).head(10)

plt.figure(figsize=(14,7))
plt.bar(cheap_large['location'], cheap_large['avg_area'])
plt.xticks(rotation=45)
plt.xlabel("Location")
plt.ylabel("Average Area")
plt.title("Cheapest Areas with Larger Properties")
plt.grid(True)
plt.show()
```
